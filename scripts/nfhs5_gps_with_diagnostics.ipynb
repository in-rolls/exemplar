```python
import geopandas as gpd
import pandas as pd

# NFHS-5 GPS
gps5 = gpd.read_file("/Users/sonalideliwala/Library/Mobile Documents/com~apple~CloudDocs/Desktop/India Data/NFHS GIS/Geographic data nfhs5/IAGE7AFL.shp")
gps5_clean = gps5[["DHSCLUST", "LATNUM", "LONGNUM", "URBAN_RURA", "ADM1NAME", "DHSREGNA"]].copy()
gps5_clean = gps5_clean[(gps5_clean["LATNUM"] != 0) & (gps5_clean["LONGNUM"] != 0)]
gps5_clean["ADM1NAME"] = gps5_clean["ADM1NAME"].str.strip()

# Filter to RJ and UP and convert to GeoDataFrame with CRS
gps5_rj_up = gps5_clean[gps5_clean["ADM1NAME"].isin(["RAJASTHAN", "UTTAR PRADESH"])].copy()
gps5_rj_up = gpd.GeoDataFrame(gps5_rj_up,
              geometry=gpd.points_from_xy(gps5_rj_up.LONGNUM, gps5_rj_up.LATNUM),
              crs="EPSG:4326")

print("NFHS clusters:", gps5_rj_up.shape)
```

```python
shrug = gpd.read_file("/Users/sonalideliwala/Library/Mobile Documents/com~apple~CloudDocs/Desktop/[RA work] Womens Reservation & Girls Aspirations/shrug-pc11-village-poly-shp/village_modified.shp")
shrug_rj_up = shrug[shrug["pc11_s_id"].isin(["08", "09"])].copy()
shrug_rj_up = shrug_rj_up.set_crs("EPSG:4326")

print("SHRUG villages:", shrug_rj_up.shape)
print(shrug_rj_up["pc11_s_id"].value_counts())
```

```python
# Reproject to meters (UTM Zone 44N covers RJ + UP well)
gps5_proj = gps5_rj_up.to_crs("EPSG:32644")
shrug_proj = shrug_rj_up.to_crs("EPSG:32644")

# Re-run spatial join with correct projection
crosswalk = gpd.sjoin_nearest(gps5_proj, shrug_proj, how="left")

crosswalk = crosswalk[["DHSCLUST", "LATNUM", "LONGNUM", "URBAN_RURA",
                        "ADM1NAME", "DHSREGNA", "tv_name",
                        "pc11_s_id", "pc11_d_id", "pc11_tv_id"]].copy()

print(crosswalk.shape)
crosswalk.head(10)
```

```python
crosswalk.to_csv("/Users/sonalideliwala/Library/Mobile Documents/com~apple~CloudDocs/Desktop/India Data/NFHS GIS/nfhs5_village_crosswalk.csv", index=False)
print("Saved!")
```

```python
women5 = pd.read_stata(
    "/Users/sonalideliwala/Library/Mobile Documents/com~apple~CloudDocs/Desktop/[RA work] Womens Reservation & Girls Aspirations/nfhs5_women/nfhs5_women_truncated.dta",
    convert_categoricals=False
)
print(women5.shape)
women5.head()
# Rename cluster variable to match crosswalk
women5 = women5.rename(columns={"v001": "DHSCLUST"})

# Merge
merged5 = women5.merge(crosswalk, on="DHSCLUST", how="left")

print(f"Women's survey rows:  {women5.shape[0]}")
print(f"Merged rows:          {merged5.shape[0]}")
print(f"Matched (non-null):   {merged5['tv_name'].notna().sum()}")
print(f"Unmatched:            {merged5['tv_name'].isna().sum()}")

merged5.head()

merged5.to_stata(
    "/Users/sonalideliwala/Library/Mobile Documents/com~apple~CloudDocs/Desktop/[RA work] Womens Reservation & Girls Aspirations/nfhs5_women/nfhs5_women_gps_merged.dta",
    write_index=False,
    version=118        # Stata 14+ format
)
print("Saved!")
```

```python
# filtering to just Rajasthan and UP
merged5_rj_up = merged5[merged5["tv_name"].notna()].copy()

print(merged5_rj_up.shape)
print(merged5_rj_up["ADM1NAME"].value_counts())
merged5_rj_up.to_stata(
    "/Users/sonalideliwala/Library/Mobile Documents/com~apple~CloudDocs/Desktop/[RA work] Womens Reservation & Girls Aspirations/nfhs5_women/nfhs5_women_gps_rj_up.dta",
    write_index=False,
    version=118
)
print("Saved!")
```

```python
# How many clusters per village?
clusters_per_village = merged5_rj_up.groupby("pc11_tv_id")["DHSCLUST"].nunique()
print(clusters_per_village.value_counts())
print(f"\nMean clusters per village: {clusters_per_village.mean():.2f}")
print(f"Villages with 2+ clusters: {(clusters_per_village > 1).sum()}")

# And how many women per village?
women_per_village = merged5_rj_up.groupby("pc11_tv_id").size()
print(women_per_village.describe())
```

```python
import pandas as pd

# Simpler file: direct village Census code → GP LGD code
village_to_gp = pd.read_csv(
    "/Users/sonalideliwala/Library/Mobile Documents/com~apple~CloudDocs/Desktop/India Data/Shrid To Local Governance Body Matching/documentation/All India Village to GP LGD codes.csv",
    encoding='latin-1'
)
print("Village-to-GP file:", village_to_gp.shape)
print(village_to_gp.columns.tolist())
village_to_gp.head()
# SHRUG-to-LGD matched file (has shrid + additional flags)
shrug_lgd = pd.read_csv(
    "/Users/sonalideliwala/Library/Mobile Documents/com~apple~CloudDocs/Desktop/India Data/Shrid To Local Governance Body Matching/data/shrug_LGD_matched.csv",
    encoding='latin-1'
)
print("SHRUG-LGD file:", shrug_lgd.shape)
print(shrug_lgd.columns.tolist())
shrug_lgd.head()
```

```python
# Step 1: Redo spatial join keeping pc11_sd_id
crosswalk = gpd.sjoin_nearest(gps5_proj, shrug_proj, how="left")

crosswalk = crosswalk[[
    "DHSCLUST", "LATNUM", "LONGNUM", "URBAN_RURA",
    "ADM1NAME", "DHSREGNA",
    "tv_name",
    "pc11_s_id", "pc11_d_id", "pc11_sd_id", "pc11_tv_id"   # now keeping pc11_sd_id
]].copy()

print(crosswalk.shape)
crosswalk.head()
```

```python
# shrug_lgd shrid2 format: "11-{state}-{district}-{subdistrict}-{village}"
# e.g. "11-01-001-00001-000001"
# We reconstruct this from SHRUG polygon fields

crosswalk["shrid2_key"] = (
    "11-" +
    crosswalk["pc11_s_id"].astype(str).str.zfill(2) + "-" +
    crosswalk["pc11_d_id"].astype(str).str.zfill(3) + "-" +
    crosswalk["pc11_sd_id"].astype(str).str.zfill(5) + "-" +
    crosswalk["pc11_tv_id"].astype(str).str.zfill(6)
)

# Verify format looks right
print(crosswalk["shrid2_key"].head(5).tolist())
# Should look like: ['11-08-099-00123-066362', ...]

# Keep only GP rows from shrug_lgd (exclude towns, ULBs etc.)
shrug_lgd_gps = shrug_lgd[
    shrug_lgd["local_body_type"] == "Gram Panchayat"
][["shrid2", "LGD_code", "local_body_name", "gp_to_urban_conversion",
   "shrid_part_of_multi_village_GP"]].copy()

shrug_lgd_gps = shrug_lgd_gps.rename(columns={
    "shrid2":           "shrid2_key",
    "LGD_code":         "gp_lgd_code",
    "local_body_name":  "gp_name"
})

# Merge onto crosswalk
crosswalk_with_gp = crosswalk.merge(shrug_lgd_gps, on="shrid2_key", how="left")

print(f"Crosswalk rows:        {crosswalk.shape[0]}")
print(f"Matched to GP:         {crosswalk_with_gp['gp_lgd_code'].notna().sum()}")
print(f"Unmatched:             {crosswalk_with_gp['gp_lgd_code'].isna().sum()}")
print(f"Match rate:            {crosswalk_with_gp['gp_lgd_code'].notna().mean():.1%}")

print("\nMatch rate by state:")
print(crosswalk_with_gp.groupby("ADM1NAME")["gp_lgd_code"].apply(
    lambda x: f"{x.notna().mean():.1%} ({x.notna().sum()} of {len(x)})"
))
```

```python
# Merge crosswalk (now with GP codes) onto women's survey
women5 = women5.rename(columns={"v001": "DHSCLUST"})

merged_final = women5.merge(
    crosswalk_with_gp[[
        "DHSCLUST", "LATNUM", "LONGNUM", "URBAN_RURA",
        "ADM1NAME", "DHSREGNA", "tv_name",
        "pc11_s_id", "pc11_d_id", "pc11_sd_id", "pc11_tv_id",
        "gp_lgd_code", "gp_name",
        "gp_to_urban_conversion", "shrid_part_of_multi_village_GP"
    ]],
    on="DHSCLUST",
    how="inner"   # inner keeps only RJ+UP women (crosswalk only has RJ+UP clusters)
)

print(f"Final dataset: {merged_final.shape}")
print(f"Unique GPs matched: {merged_final['gp_lgd_code'].nunique()}")
print(merged_final["ADM1NAME"].value_counts())

# Save
merged_final.to_stata(
    "/Users/sonalideliwala/Library/Mobile Documents/com~apple~CloudDocs/Desktop/[RA work] Womens Reservation & Girls Aspirations/nfhs5_women/nfhs5_women_gp_final.dta",
    write_index=False,
    version=118
)
print("Saved!")
```

```python
# Check what local_body_types exist in shrug_lgd
print("All local_body_type values:")
print(shrug_lgd["local_body_type"].value_counts().head(20))

# Check match rate broken down by urban/rural cluster
print("\nCluster urban/rural breakdown:")
print(crosswalk_with_gp.groupby("URBAN_RURA")["gp_lgd_code"].apply(
    lambda x: f"{x.notna().mean():.1%} ({x.notna().sum()} of {len(x)})"
))

# Check unmatched clusters - what do their shrid2 keys look like?
unmatched = crosswalk_with_gp[crosswalk_with_gp["gp_lgd_code"].isna()]
print(f"\nUnmatched clusters: {len(unmatched)}")
print("Sample unmatched shrid2 keys:")
print(unmatched["shrid2_key"].head(10).tolist())

# Check if those shrid2 keys exist in shrug_lgd at all (regardless of type)
unmatched_keys = unmatched["shrid2_key"].tolist()
found_in_shrug = shrug_lgd[shrug_lgd["shrid2"].isin(unmatched_keys)]
print(f"\nUnmatched keys found in shrug_lgd (any type): {len(found_in_shrug)}")
print(found_in_shrug["local_body_type"].value_counts())
```

```python
print("Match rate by state AND urban/rural:")
print(crosswalk_with_gp.groupby(["ADM1NAME", "URBAN_RURA"])["gp_lgd_code"].apply(
    lambda x: f"{x.notna().mean():.1%} ({x.notna().sum()} of {len(x)})"
))

# Check how many of unmatched rural clusters have OG village codes (start with 8)
rural_unmatched = crosswalk_with_gp[
    (crosswalk_with_gp["gp_lgd_code"].isna()) & 
    (crosswalk_with_gp["URBAN_RURA"] == "R")
]
og_flag = rural_unmatched["pc11_tv_id"].astype(str).str.startswith("8")
print(f"\nUnmatched rural clusters with OG village codes (start with 8): {og_flag.sum()}")
print(f"Unmatched rural clusters - genuine LGD gap:                    {(~og_flag).sum()}")
```

```python
# Are unmatched UP rural clusters systematically different?
# Compare key characteristics between matched and unmatched

up_rural = merged_final[
    (merged_final["ADM1NAME"] == "UTTAR PRADESH") &
    (merged_final["URBAN_RURA"] == "R")
].copy()

up_rural["matched"] = up_rural["gp_lgd_code"].notna().astype(int)

# Compare on observable characteristics
check_vars = ["v190",   # wealth index
              "v149",   # educational attainment
              "v012",   # age
              "v025",   # urban/rural (should all be rural)
              "v744a",  # wife beating attitudes
              "v743a"]  # decision-making autonomy

print("Matched vs Unmatched UP Rural clusters - mean comparison:")
print(up_rural.groupby("matched")[check_vars].mean().T.rename(
    columns={0: "Unmatched", 1: "Matched"}
))

# Also check geographic spread - are unmatched in particular districts?
print("\nUnmatched UP rural by district (DHSREGNA):")
up_rural_unmatched = up_rural[up_rural["matched"] == 0]
print(up_rural_unmatched["DHSREGNA"].value_counts().head(15))
```
